```python
import pandas as pd
import matplotlib.pyplot as plt

plt.rc('font', family='NanumBarunGothic')
```

```python
df_temp = pd.read_csv('/content/seoul_temp.csv')
# df_temp
# df_temp.head(5) # 상위 n행 출력
# df_temp.tail(7) # 하위 n행 출력
# df_temp.info()  # 인덱스와 컬럼정보
# df_temp.describe()  # 컬럼별 통계량
# df_temp.shape   # 행열의 크기 [0]:행 [1]:열
# df_temp.columns # 컬럼 이름 확인
df_temp.set_index('연도', inplace=True) # 인덱스 변경
```

```python
plt.figure(figsize=(20,20))

plt.plot(df_temp.index, df_temp['평균기온'])
plt.plot(df_temp.index, df_temp['최저기온'])
plt.plot(df_temp.index, df_temp['최고기온'])

plt.title('서울시연간기온')
plt.xlabel('연도')
plt.ylabel('기온')

plt.plot(color='red') # 색 변경
plt.grid()    # 격자 추가
plt.plot(marker='o')  # 마커 추가
plt.show()
```
